# Pairs Trading Strategy 

In this Project, I am attempting to formulate a Pairs Trading Strategy on two Japanese shipping companies stocks, Nippon Yusen and Shosen Mitsui, 9101.T and 9104.T respectively. This is a mean reversion market-neutral strategy that capitalizes divergence of spread between the two pairs.

```python
def trade(S1, S2, window1, window2, money):
    # If window length is 0, algorithm doesn't make sense, so exit
    if (window1 == 0) or (window2 == 0):
        return 0
    # Compute rolling mean and rolling standard deviation
    ratios = S1/S2
    ma1 = ratios.rolling(window=window1,
                               center=False).mean()
    ma2 = ratios.rolling(window=window2,
                               center=False).mean()
    std = ratios.rolling(window=window2,
                        center=False).std()
    zscore = (ma1 - ma2)/std
    # Simulate trading
    # Start with no money and no positions
    numTrades = 0
    countS1 = 0
    countS2 = 0
    pnl = []
    for i in range(len(ratios)):
        # Sell short if the z-score is > 1
        if zscore[i] > 1.25:
            money += S1[i] - S2[i] * ratios[i]
            countS1 -= 1
            countS2 += ratios[i]
        # Buy long if the z-score is < 1
        elif zscore[i] < -1.25:
            money -= S1[i] - S2[i] * ratios[i]
            countS1 += 1
            countS2 -= ratios[i]
        # Clear positions if the z-score between -.5 and .5
        elif abs(zscore[i]) < 0.5:
            money += countS1*S1[i] + S2[i] * countS2
            countS1 = 0
            countS2 = 0
            numTrades += 1
#         print('Z-score: '+ str(zscore[i]), countS1, countS2, S1[i] , S2[i])
        pnl.append([ratios.index[i], money])
    result = pd.DataFrame(pnl, columns=["Date", "pnl"])
    result = result.set_index("Date")
    return result, numTrades
```

```python
result = trade(d_9104, d_9107, 5, 30, 1000)
pnl = result[0]
```

```python
def risk(pnl, money):
    end = 0 
    start = 0 
    s = pd.Series(dtype=object)
    s.loc['Start Date'] = pnl.index[0]
    s.loc['End Date'] = pnl.index[-1]
    s.loc['Duration'] = str(len(pnl)) + ' Days'
    s.loc['Start Equity'] = '$' + str(money)
    s.loc['Final Equity'] = '$' + str(round(pnl.tail(1).values[0][0]))
    returnVal = (pnl.tail(1).values[0][0]-money)/money*100
    s.loc['Return'] = str(round(returnVal, 1)) + '%'
    #s.loc['Vol'] = pnl.std()
    s.loc['Number of Trades'] = result[1]

    end = (pnl.cummax()- pnl).idxmax()
    start = pnl.cummax()[:end[0]].idxmax()
    
    dd = str(round((pnl.loc[start].iloc[0] - pnl.loc[end].iloc[0])[0]/pnl.loc[start].iloc[0][0], 5)*100) + '%'
    s.loc['Max Drawdown'] = dd
    s.loc['Drawdown Duration'] = str(len(pnl[start[0]:end[0]])) + ' Days'
    return s
```

```python
def trade_2(S1, S2, window1, window2, money, t_cost):
    # If window length is 0, algorithm doesn't make sense, so exit
    if (window1 == 0) or (window2 == 0):
        return 0
    # Compute rolling mean and rolling standard deviation
    ratios = S1/S2
    ma1 = ratios.rolling(window=window1,
                               center=False).mean()
    ma2 = ratios.rolling(window=window2,
                               center=False).mean()
    std = ratios.rolling(window=window2,
                        center=False).std()
    zscore = (ma1 - ma2)/std
    # Simulate trading
    # Start with no money and no positions
    numTrades = 0
    winTrades = 0
    countS1 = 0
    countS2 = 0
    unrealized_gain = 0
    unr_pnl = []
    pnl = []
    for i in range(len(ratios)):
        # Sell short if the z-score is > 1
        if zscore[i] > 1.25:
            countS1 -= 1
            countS2 += ratios[i]
            numTrades += 1
            money = money * (1-t_cost)
        # Buy long if the z-score is < 1
        elif zscore[i] < -1.25:
            countS1 += 1
            countS2 -= ratios[i]
            numTrades += 1
            money = money * (1-t_cost)
        # Clear positions if the z-score between -.5 and .5
        elif abs(zscore[i]) < 0.5:
            if(countS1 != 0 and countS2 != 0):
                numTrades += 2
                
            prev_money = money
            money += countS1*S1[i] + S2[i] * countS2
            money = money * (1-t_cost)
            
            if money > prev_money:
                winTrades += 1
            countS1 = 0
            countS2 = 0
#         print('Z-score: '+ str(zscore[i]), countS1, countS2, S1[i] , S2[i]

        else: 
            unrealized_gain = (countS1*S1[i] + S2[i] * countS2)
            capital = unrealized_gain + money
  
        #print('S1: ' + str(countS1) + ' ' + str(S1[i]))
        #print('S2: ' + str(countS2) + ' ' + str(S2[i]))
        #print('unrealized gain: ' + str(countS1*S1[i] + S2[i] * countS2))
        
        
        pnl.append([ratios.index[i], money])
        unr_pnl.append([ratios.index[i], capital])
        
        
    result = pd.DataFrame(pnl, columns=["Date", "pnl"])
    result = result.set_index("Date")
    
    unr_result = pd.DataFrame(unr_pnl, columns=["Date", "unrealized_pnl"])
    unr_result = unr_result.set_index("Date")
    
    result = pd.concat([result, unr_result], axis=1)
    
    #print('countS1: ' + str(countS1))
    #print('countS2: ' + str(countS2))
    return result, numTrades
```

```python
result1 = trade(d_9104[3000:], d_9107[3000:], 5, 30, 1000)
pnl1 = result1[0]
result2 = trade_2(d_9104[3000:], d_9107[3000:], 5, 30, 1000, 0.002)
pnl2 = result2[0]
```
